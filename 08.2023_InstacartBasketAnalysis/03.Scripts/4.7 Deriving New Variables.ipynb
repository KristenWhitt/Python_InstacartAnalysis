```python
#create function for price range label
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#apply price_label function and put it in new column called price_range
df_sub['price_range'] = df_sub.apply(price_label, axis = 1)
```

```python
#check frequency of labels in price_range
df_sub['price_range'].value_counts()
```

```python
#use .loc() to address warning above
df_sub.loc[df_sub['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
#use .loc() to address warning above
df_sub.loc[(df_sub['prices'] <= 15) & (df_sub['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

## Create new column busiest_days with top two busiest days and bottom two busiest days 

### busiest days are 0 & 1, least busy are 3 & 4 

```python
#create empty result_1 list for busiest_days 
result_1 = []
```

```python
#two busiest days 0 & 1, two slowest days 3 & 4
#create for loop to determine what type of day products purchased on
for value in df_ords_prods_merged['order_day_of_week']:
  if (value == 0) | (value == 1):
    result_1.append('Busiest days')
  elif (value == 3) | (value == 4):
    result_1.append('Least busy days')
  else:
    result_1.append('Regularly busy')
```

```python
#check subset of output from for loop
result_1[:100]
```

```python
#add new column to merged dataframe
df_ords_prods_merged['busiest_days'] = result_1
```

```python
#check frequencies to compare to frequencies from order_day_of_week
df_ords_prods_merged['busiest_days'].value_counts()
```

### busiest days count matches sum of order day 0 and order day 1: 11,864,412
### least busy days count matches sum of order day 3 and order day 4: 7,624,336

## Create new column for busiest period of day with labels 'most orders', 'average orders', 'fewest orders'

```python
#check frequencies to determine hours of day with highest, average, and lowest orders
df_ords_prods_merged['order_hour_of_day'].value_counts(sort = True)
```

```python
#assume frequencies over 2 million app will crash. hours 9,10,11,12,13,14,15,16,17 over 2 million
#assume frenquencies under 100 thousand considered not busy. hours 2,3,4,5 under 100 thousand
#assume all other frequencies considered average
#create results list for new column
result_2 = []
```

```python
#create for loop for labels for hour of day
for value in df_ords_prods_merged['order_hour_of_day']:
    if (value >=2) and (value <=5):
        result_2.append('fewest orders')
    elif (value >=9) and (value <=17):
        result_2.append('most orders')
    else:
        result_2.append('average orders')
```

```python
#check subset of result_2 for output
result_2[:100]
```

```python
#add new column to dataframe and check frequencies of new column
df_ords_prods_merged['busiest_period_of_day'] = result_2
df_ords_prods_merged['busiest_period_of_day'].value_counts()
```

### value of most orders matches sum of hours 9-17: 23,205,725
### value of fewest orders matches sum of hours 2-5: 261,859
### value of average orders matches sum of remaining hours: 8,937,275
